````python
def bnd2cnt(x):
    """xyxy to cxcywh"""
    x0, y0, x1, y1 = x.unbind(-1)

    w = x1 - x0
    h = y1 - y0
    return x0.item(), y0.item(), w.item(), h.item()

def draw_mask(mask, box, score, savefp=None):
    """Draw a single mask with bounding box and score."""
    fig, ax = plt.subplots()

    # show mask
    ax.imshow(mask, cmap='gray')

    # draw bounding box
    x, y, w, h = bnd2cnt(box)
    rect = plt.Rectangle((x, y), w, h, fill=False, edgecolor='red', linewidth=3)
    ax.add_patch(rect)
    ax.text(x*(1-0.2), y*(1-0.05), 'event:\n {:.2f}'.format(score), fontsize=20, color='yellow', fontweight='bold')
    plt.axis('off')
    if savefp is not None:
        fig.savefig(savefp, dpi=600, bbox_inches='tight', pad_inches=0)

@torch.no_grad()
def draw_masks(fp, name, savefp=None):
    im = Image.open(os.path.join(fp, name+'.png')).convert('RGB').resize((256, 256), Image.ANTIALIAS)
    ts = transforms.ToTensor()(im)
    out = model([ts.cuda()])[0]
    if savefp is not None:
        mkdirs(savefp)
        for i in range(len(out['scores'])):
            draw_mask(out['masks'][i, 0].detach().cpu().numpy(), out['boxes'][i], out['scores'][i].item(), savefp=f'{savefp}/{name}_mask_{i}.png')
    else:
        return out['masks']


    
def convert_matlab_time(matlab_time):
    try:
        return (datetime.fromordinal(int(matlab_time)) + timedelta(days=matlab_time%1) - timedelta(days = 366)).date()
    except:
        return np.nan
````

## Visualize GRs

````python
def vis_GR(df, name, index, vmax, savefp='figures'):
    mask = reshape_mask(draw_masks('datasets/img', name)[index,0].detach().cpu().numpy() >= 0.50, df[name].shape)
#     pos = np.where(mask)
#     ymin = np.min(pos[1])
#     ymax = np.max(pos[1])
#     dps = [float(dp) for dp in df.columns]
#     dp_start = dps[ymin]
#     dp_end = dps[ymax]
    _, x_pred, y_pred  = get_GR(df[name], mask, 2.72, 25.2, savefp=savefp, tm_res=10, vmax=vmax)
    return [x_pred, y_pred*1e-9]
````

````python
dfs = []
namess = []
GRs = []
````

### $\mathrm{V\ddot{a}rri\ddot{o}}$

````python
df = pd.read_csv('datasets/varrio_gr.csv', parse_dates=[0], index_col=0)

names = ['2005-04-29', '2015-10-28', '2015-10-28', '2017-05-21']
for idx, name in enumerate(names):
    dfs.append(df[name])
    namess.append(name)
    if idx == 2:
        GRs.append(vis_GR(df[name], name, 1, 1e4))
    else:
        GRs.append(vis_GR(df[name], name, 0, 1e4))
len(GRs), len(dfs), len(namess)
````
